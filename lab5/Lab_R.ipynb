```r
library("ggplot2")
```

```r
j_results = read.csv("jtt_lab3.csv")
j_results
```

```r
c_results = read.csv("ctt_lab3.csv")
c_results
```

```r
co_results = read.csv("cott_lab3.csv")
co_results
```

```r
avg_j_results = aggregate( time ~ f.x.:size, data=j_results, FUN=mean)
avg_j_results$std_dev = aggregate( time ~ f.x.:size, data=j_results, FUN=sd)$time
avg_j_results
```

```r
avg_c_results = aggregate( time ~ f.x.:size, data=c_results, FUN=mean)
avg_c_results$std_dev = aggregate( time ~ f.x.:size, data=c_results, FUN=sd)$time
avg_c_results
```

```r
avg_co_results = aggregate( time ~ f.x.:size, data=co_results, FUN=mean)
avg_co_results$std_dev = aggregate( time ~ f.x.:size, data=co_results, FUN=sd)$time
avg_co_results
```

```r
ggplot(avg_j_results, aes(size,time,color=f.x.)) + 
geom_errorbar(aes(ymin=time-std_dev, ymax=time+std_dev), width=.1, color="black") +
geom_smooth(method="lm", se=TRUE, fill=NA, formula=y ~ poly(x, 3, raw=TRUE)) +
geom_point() + 
ylab("Matrix multiplication time in seconds") +
xlab("Matrix_size")
```

```r
ggplot(avg_c_results, aes(size,time,color=f.x.)) + 
geom_errorbar(aes(ymin=time-std_dev, ymax=time+std_dev), width=.1, color="black") +
geom_smooth(method="lm", se=TRUE, fill=NA, formula=y ~ poly(x, 3, raw=TRUE)) +
geom_point() + 
ylab("Matrix multiplication time in seconds") +
xlab("Matrix_size")
```

```r
ggplot(avg_co_results, aes(size,time,color=f.x.)) + 
geom_errorbar(aes(ymin=time-std_dev, ymax=time+std_dev), width=.1, color="black") +
geom_smooth(method="lm", se=TRUE, fill=NA, formula=y ~ poly(x, 3, raw=TRUE)) +
geom_point() + 
ylab("Matrix multiplication time in seconds") +
xlab("Matrix_size")
```

```r
avg_results <- merge(merge(avg_j_results,avg_c_results, all.x=TRUE, all.y=TRUE), avg_co_results, all.x=TRUE, all.y=TRUE) 


ggplot(avg_results, aes(size,time,color=f.x.)) + 
geom_errorbar(aes(ymin=time-std_dev, ymax=time+std_dev), width=.1, color="black") +
geom_smooth(method="lm", se=TRUE, fill=NA, formula=y ~ poly(x, 3, raw=TRUE)) +
geom_point()


xlab("Matrix_size") + 
ylab("Computation time")
```
